```python
from sklearn.preprocessing import StandardScaler

# Initialize scaler
scaler = StandardScaler()

# Scale features
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

# Initialize models
models = {
    "Linear Regression": LinearRegression(),
    "Ridge Regression": Ridge(alpha=1.0),
    "Decision Tree": DecisionTreeRegressor(random_state=42)
}

# To store results
results = []

for name, model in models.items():
    if name in ["Linear Regression", "Ridge Regression"]:
        # Use scaled data
        model.fit(X_train_scaled, y_train)
        y_train_pred = model.predict(X_train_scaled)
        y_test_pred = model.predict(X_test_scaled)
    else:
        # Use original data for tree
        model.fit(X_train, y_train)
        y_train_pred = model.predict(X_train)
        y_test_pred = model.predict(X_test)

    # Compute metrics
    train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
    test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))
    test_mae = mean_absolute_error(y_test, y_test_pred)

    results.append([name, train_rmse, test_rmse, test_mae])
```

```python
import pandas as pd

comparison = pd.DataFrame(results, columns=["Model", "Train RMSE", "Test RMSE", "Test MAE"])
comparison
```

```python
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np
import pandas as pd

# Models
models = {
    "Linear Regression": LinearRegression(),
    "Ridge Regression": Ridge(alpha=1.0),
    "Decision Tree": DecisionTreeRegressor(random_state=42)
}

# To store results
results = []

for name, model in models.items():
    if name in ["Linear Regression", "Ridge Regression"]:
        # Use scaled data
        model.fit(X_train_scaled, y_train)
        y_train_pred = model.predict(X_train_scaled)
        y_test_pred = model.predict(X_test_scaled)
    else:
        # Decision Tree does not need scaling
        model.fit(X_train, y_train)
        y_train_pred = model.predict(X_train)
        y_test_pred = model.predict(X_test)

    # Compute metrics
    train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
    test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))
    test_mae = mean_absolute_error(y_test, y_test_pred)

    results.append([name, train_rmse, test_rmse, test_mae])

# Comparison Table
comparison = pd.DataFrame(results, columns=["Model", "Train RMSE", "Test RMSE", "Test MAE"])
comparison
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
y_test_pred = rf.predict(X_test)

train_rmse = np.sqrt(mean_squared_error(y_train, rf.predict(X_train)))
test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))
test_mae = mean_absolute_error(y_test, y_test_pred)

print("Random Forest")
print("Train RMSE:", train_rmse)
print("Test RMSE:", test_rmse)
print("Test MAE:", test_mae)
```

```python
from sklearn.ensemble import RandomForestRegressor

# Train Random Forest
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
y_train_pred_rf = rf.predict(X_train)
y_test_pred_rf = rf.predict(X_test)

# Compute metrics
train_rmse_rf = np.sqrt(mean_squared_error(y_train, y_train_pred_rf))
test_rmse_rf = np.sqrt(mean_squared_error(y_test, y_test_pred_rf))
test_mae_rf = mean_absolute_error(y_test, y_test_pred_rf)

# Add to previous results
results.append(["Random Forest", train_rmse_rf, test_rmse_rf, test_mae_rf])

# Make final comparison table
comparison = pd.DataFrame(results, columns=["Model", "Train RMSE", "Test RMSE", "Test MAE"])
comparison
```

## Model Comparison  Bias & Variance Analysis

Linear Regression and Ridge Regression show **underfitting (high bias)** because their training and testing RMSE values are almost equal and relatively high. These models assume linear relationships and fail to capture the complex non-linear patterns in the data.

The Decision Tree model shows **overfitting (high variance)** as it achieves zero training RMSE but much higher test RMSE, meaning it memorizes training data and does not generalize well.

Random Forest reduces overfitting by averaging predictions from multiple decision trees. This results in significantly lower test RMSE and MAE compared to a single Decision Tree.

Real-world machine learning challenges such as **non-linear relationships, noisy features, and outliers** affect model performance. Ensemble methods and regularization help address these issues.
